```python
import pandas as pd
```

```python
rent = pd.read_csv('lloguerbcn.csv')
rent.tail()
```

We will clean data, from the first file we only want data from the 3rd trimester and expressed in absolute price, not in euros/m2. We only need the columns 'districte', 'barri' and 'preu'. 

```python
cleanrent = rent[rent['Trimestre'] == 3]
cleanrent = cleanrent[cleanrent['Preu'] > 100]
cleanrent = cleanrent.drop('Lloguer_mitja', 1)
cleanrent = cleanrent.drop('Trimestre', 1)
cleanrent = cleanrent.drop('Codi_Districte', 1)
cleanrent = cleanrent.drop('Codi_Barri', 1)
cleanrent = cleanrent.drop('Any', 1)

cleanrent
```

For the first analysis we only need 'Districte' and 'Price'.

```python
districts = cleanrent.drop('Nom_Barri', 1)
districts = round(districts.groupby(['Nom_Districte']).mean(), 2)
districts
```
